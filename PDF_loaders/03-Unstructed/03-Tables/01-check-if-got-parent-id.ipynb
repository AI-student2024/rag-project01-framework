```python
from unstructured.partition.pdf import partition_pdf
```

```python
# 提取 PDF 的表格内容
elements = partition_pdf(
    filename="/home/huangjia/Documents/12_SAP_Cycle17/data/Boeing-Sustainability-Report/1_page/Boeing-Sustainability-Report_074.pdf",
    # strategy="hi_res",
    strategy="ocr_only",
    infer_table_structure=True,  # 启用表格结构推断
)
```

```python
# 查找有没有parent_id的元素
for element in elements:
    if hasattr(element.metadata, 'parent_id'):
        print(element.metadata.parent_id)
```
