```python
elements = partition(
    "/home/huangjia/Documents/12_SAP_Cycle17/data/Boeing-Sustainability-Report/1_page/Boeing-Sustainability-Report_067.pdf",
    strategy="hi_res",
    split_pdf_page=True,
    split_pdf_allow_failed=True,
    split_pdf_concurrency_level=15
)
```

```python
import json

# Load and partition the PDF
# elements = partition("path/to/your/document.pdf")

# Convert elements to a list of dictionaries
element_dicts = [elem.to_dict() for elem in elements]

# Save as JSON
with open("output.json", "w") as f:
    json.dump(element_dicts, f, indent=2)
```
